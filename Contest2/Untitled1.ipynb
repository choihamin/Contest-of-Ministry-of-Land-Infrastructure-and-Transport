```python
import os
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
from tqdm import tqdm
```

```python
import urllib.request

url = "http://openapi.kepco.co.kr/service/EvInfoServiceV2/getEvSearchList"
key1 = "QD%2F1soJk0R03qtRk4Xzsz7%2FqQ9G0hCAx7Biw1ycmZ3GJMS%2BVS2PkNqedYdjUgR22z240xP%2FsTIweFQI6QdJcDA%3D%3D"
key2 = "QD/1soJk0R03qtRk4Xzsz7/qQ9G0hCAx7Biw1ycmZ3GJMS+VS2PkNqedYdjUgR22z240xP/sTIweFQI6QdJcDA=="
i = 0
pn = 1
num_of_row = 9999
req = urllib.request.urlopen('{}?ServiceKey={}&pageNo={}&numOfRows={}'.format(url, key1, pn, num_of_row))
soup = BeautifulSoup(req, "html.parser")

df = pd.DataFrame(columns=['addr', 'charTp', 'cpId', 'cpNm', 
                           'cpStat', 'cpTp', 'csId', 'csNm',
                           'lat','longi','statUpdateDatetime'])
items = soup.find("items")
for item in tqdm(items.findAll("item")):
    df.loc[i] = [item.addr.text, item.chargetp.text, item.cpid.text, item.cpnm.text,
                 item.cpstat.text, item.cptp.text, item.csid.text, item.csnm.text,
                 item.lat.text, item.longi.text, item.statupdatedatetime.text]
    i += 1
df
```
