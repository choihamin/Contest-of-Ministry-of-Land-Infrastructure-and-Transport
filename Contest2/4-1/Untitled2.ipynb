```python
import pandas as pd
file = pd.read_csv('report.txt', delimiter = '\t', thousands =',')
```

```python
file
```

```python
file['직장'] = file['주택용'] + file['일반용'] + file['교육용']
```

```python
file
```

```python
file2 = file[file['기간']==2015]
file2
```

```python
file2 = file2[['자치구', '주택용']].iloc[1:26,]
file2
```

```python
import folium
import pandas as pd
import json
import re
```

```python
geo_json='https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'

file2=file2.sort_values(by='자치구')
file2['자치구']=file2['자치구'].apply(lambda x: re.compile('[가-힣]+').findall(x)[0])

m=folium.Map(
    location=[37.566345,126.977893],
    tiles='Stamen Terrain',
    zoom_start=11
)

folium.Choropleth(
    geo_data=geo_json,
    name='choropleth',
    data=file2,
    columns=['자치구','주택용'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2
).add_to(m)
```

```python
m
```
